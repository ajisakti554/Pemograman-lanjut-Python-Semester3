Buatlah program yang menerima bilangan N dan menuliskan bilangan 10x terkecil yang lebih dari N

```python
#Buatlah program yang menerima bilangan N dan menuliskan bilangan 10x terkecil yang lebih dari N.

def cari_10pangkatx(n):
  x = 0
  pangkat10 = 1
  while pangkat10 <= n:
    x += 1
    pangkat10 = 10 ** x
  return pangkat10

n = int(input("Masukkan bilangan N: "))
hasil = cari_10pangkatx(n)
print("Bilangan 10 pangkat x terkecil yang lebih dari", n, "adalah:", hasil)
```

buatlah program yang menerima bilangan N dan membuat pola segitiga angka dengan dasar N

```python
def buat_segitiga_angka(n):
  for i in range(1, n + 1):
    for j in range(i):
      print(j + 1, end=" ")
    print()

n = int(input("Masukkan bilangan N: "))
buat_segitiga_angka(n)
```

MEMBUAT KALKULATOR SEDERHANA

```python
def kalkulator():
    """Fungsi untuk membuat kalkulator sederhana."""

    # Meminta input dari pengguna
    angka1 = float(input("Masukkan angka pertama: "))
    operator = input("Masukkan operator (+, -, *, /): ")
    angka2 = float(input("Masukkan angka kedua: "))

    # Melakukan perhitungan berdasarkan operator
    if operator == "+":
        hasil = angka1 + angka2
    elif operator == "-":
        hasil = angka1 - angka2
    elif operator == "*":
        hasil = angka1 * angka2
    elif operator == "/":
        if angka2 == 0:
            print("Error: Pembagian dengan nol tidak diizinkan.")
            return
        hasil = angka1 / angka2
    else:
        print("Error: Operator tidak valid.")
        return

    # Menampilkan hasil
    print("Hasil :",hasil)

# Memanggil fungsi kalkulator
    kalkulator()
```

Problem (2) Bilangan Ganjil

```python
def cek_ganjil_genap(awal, akhir):
     """Fungsi untuk mengecek bilangan ganjil genap dalam range tertentu."""
     for angka in range(awal, akhir + 1):
       if angka % 2 == 0:
         print('bilangan genap',angka)
       else:
         print('bilangan ganjil',angka)

# Contoh penggunaan:
awal = 1
akhir = 10
cek_ganjil_genap(awal, akhir)
```

- Interpretasi coding: menggunakan perulangan for dengan syarat %2==0 (bilangan genap) jika sebaliknya maka bilangan ganjil

- Interpretasi Hasil:
1.Selalu berakhir dengan 1, 3, 5, 7, atau 9.
Penjumlahan dua bilangan ganjil menghasilkan bilangan genap.
Penjumlahan satu bilangan genap dan satu bilangan ganjil menghasilkan bilangan ganjil.

2.Bilangan Genap:

Selalu berakhir dengan 0, 2, 4, 6, atau 8.
Penjumlahan dua bilangan genap menghasilkan bilangan genap.
Penjumlahan dua bilangan ganjil juga menghasilkan bilangan genap.
Penjumlahan satu bilangan genap dan satu bilangan ganjil menghasilkan bilangan ganjil.
